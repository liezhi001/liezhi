```python
import numpy as np

import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y, label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Line Plot Example')
plt.legend()
plt.show()
```
