```python
import numpy as np

import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y, label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Line Plot Example')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 创建图形
plt.figure()

# 绘制正弦函数
plt.plot(x, y, label='sin(x)')

# 添加标签和标题
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Line Plot Example')

# 添加网格线
plt.grid(True)

# 添加图例
plt.legend()

# 显示图形
plt.show()

# 保存图像到文件
plt.savefig('line_plot_example.png')
```
